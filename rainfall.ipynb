```python
#관측월별 계절변수
def get_season(month):
    if month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    elif month in [9, 10, 11]:
        return 'autumn'
    
train['season_fc'] = train['rainfall_train.fc_month'].apply(get_season)
```

```python
#관측계절별 hist
grouped_season_fc = train.groupby('season_fc')
vv_means = grouped_season_fc['rainfall_train.vv'].mean()

plt.figure(figsize=(10, 6))
vv_means.plot(kind='bar', color='skyblue')
plt.title('Average Rainfall (vv) by Time Interval (dh)')
plt.xlabel('Time Interval (dh)')
plt.ylabel('Average Rainfall (vv)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 각 원소에 대해 계급을 매기는 함수
def grade(value):
    if value < 0.1:
        return 0
    elif 0.1 <= value < 0.2:
        return 1
    elif 0.2 <= value < 0.5:
        return 2
    elif 0.5 <= value < 1.0:
        return 3
    elif 1.0 <= value < 2.0:
        return 4
    elif 2.0 <= value < 5.0:
        return 5
    elif 5.0 <= value < 10.0:
        return 6
    elif 10.0 <= value < 20.0:
        return 7
    elif 20.0 <= value < 30.0:
        return 8
    else:
        return 9

train['y_pred'] = y_pred
train['pred_class_interval'] = train['y_pred'].apply(grade)
def compare_cols(row):
    if row['rainfall_train.class_interval'] == 0 and row['pred_class_interval'] == 0:
        return 0
    elif row['rainfall_train.class_interval'] == row['pred_class_interval']:
        return 1
    else:
        return 2

# apply 함수를 사용하여 새로운 열 추가
train['Match'] = train.apply(compare_cols, axis=1)
train['Match'].value_counts()
```

```python
# 파이프라인 정의
def custom_predict(model, X):
    predictions = model.predict(X)
    
    # 누적 확률 컬럼 인덱스 (변환 후 배열 기준)
    cumulative_prob_indices = [2, 3, 4, 5, 6, 7, 8, 9, 10]
    
    # 누적 확률 컬럼들이 모두 0인지 확인
    zero_cumulative_prob_indices = np.all(X[:, cumulative_prob_indices] == 0, axis=1)
    predictions[zero_cumulative_prob_indices] = 0
    
    return predictions

def sqrt_transform(X):
    return np.sqrt(X)

pipeline = Pipeline([
    ('preprocessing', ColumnTransformer([
        ('sqrt_transform', FunctionTransformer(func=sqrt_transform), ['dh', 'ef', 'v01', 'v02', 'v03', 'v04', 
            'v05', 'v06', 'v07', 'v08', 'v09'])
    ], remainder='drop')),
    ('model', RandomForestRegressor(n_estimators=100, random_state=42))
])

# 학습 데이터
X_train = train[['dh', 'ef', 'v01', 'v02', 'v03', 'v04', 'v05', 'v06', 'v07', 'v08', 'v09']]
y_train = train['vv']

# 성능 평가를 위한 교차 검증 설정
scoring = {
    'r2': make_scorer(r2_score),
    'mse': make_scorer(mean_squared_error),
    'mae': make_scorer(mean_absolute_error)
}

# 교차 검증을 통한 성능 평가
scores = cross_validate(pipeline, X_train, y_train, cv=5, scoring=scoring, return_train_score=True)

# 성능 지표 출력
print(f"R^2 (Train): {np.mean(scores['train_r2'])}")
print(f"R^2 (Test): {np.mean(scores['test_r2'])}")
print(f"MSE (Train): {np.mean(scores['train_mse'])}")
print(f"MSE (Test): {np.mean(scores['test_mse'])}")
print(f"MAE (Train): {np.mean(scores['train_mae'])}")
print(f"MAE (Test): {np.mean(scores['test_mae'])}")

# 모델 학습
pipeline.fit(X_train, y_train)

# 예측
# 테스트 데이터 예측
X_test = test[['dh', 'ef', 'v01', 'v02', 'v03', 'v04', 'v05', 'v06', 'v07', 'v08', 'v09']]
predictions = custom_predict(pipeline.named_steps['model'], pipeline.named_steps['preprocessing'].transform(X_test))
print(predictions)

```
